```python
dataset_basket = pd.read_csv('data/customer_basket.csv').set_index('customer_id')
```

```python
cluster_1_basket = dataset_basket[dataset_basket.index.isin(cluster_1.index)]
cluster_2_basket = dataset_basket[dataset_basket.index.isin(cluster_2.index)]
cluster_3_basket = dataset_basket[dataset_basket.index.isin(cluster_3.index)]
cluster_4_basket = dataset_basket[dataset_basket.index.isin(cluster_4.index)]
cluster_5_basket = dataset_basket[dataset_basket.index.isin(cluster_5.index)]
cluster_6_basket = dataset_basket[dataset_basket.index.isin(cluster_6.index)]
cluster_7_basket = dataset_basket[dataset_basket.index.isin(cluster_7.index)]
cluster_8_basket = dataset_basket[dataset_basket.index.isin(cluster_8.index)]
cluster_9_basket = dataset_basket[dataset_basket.index.isin(cluster_9.index)]
cluster_1_basket
```

----

To understand which promotions and offers can be made for **each cluster of customers** we applied **Apriori algorithm**:

## Cluster 1 - **Vegan People**

```python
te = TransactionEncoder()
te_fit = te.fit(cluster_1_basket['list_of_goods']).transform(cluster_1_basket['list_of_goods'])
dataset_cluster_1_basket = pd.DataFrame(te_fit, columns=te.columns_)

dataset_cluster_1_basket
```

## Apriori algorithm

'support' will tell us the most popular items that are bought by the customers of this cluster

Below it we can see the item(s) that is(are) bought together with the most popular items calculated by 'support' (sorted by the higher confidence to the lower)

```python
frequent_itemsets = apriori(dataset_cluster_1_basket, min_support=0.05, use_colnames=True)
display(frequent_itemsets.sort_values(by='support', ascending=False))

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6, num_itemsets=100)
rules = rules.sort_values(by='confidence', ascending=False)

display(rules.head(50).sort_values(by='confidence', ascending=False))
```

Associations with more than 1 item:

```python
frequent_itemsets_2 = apriori(
    dataset_cluster_1_basket, min_support=0.02, use_colnames=True
    )

rules_2 = association_rules(frequent_itemsets_2,
                                  metric="confidence",
                                  min_threshold=0.2, num_itemsets=100)

rules_2.sort_values(by='lift', ascending=False).head(10)
```

## **Cupons  and Promotions Strategy**

Based on the customer segmentation and association rules analysis, the following targeted coupon and promotion strategies are proposed to enhance customer engagement and sales:

- `Buy 1, Get 1 Free on Oil`: Many clusters frequently purchase oil, making it the top product for promotional offers.

- `10% Discount on Mixed Vegetables`: Customers buying four different vegetables receive a 10% discount, supporting the “Vegan People” cluster.

- `Monthly 20% Discount on Selected Vegetables`: Each month, a different vegetable is offered at 20% off, encouraging repeat purchases and appealing to vegan customers.

- `Student Card Discount on Alcoholic Beverages`: Students receive a 30% discount on alcoholic drinks purchases over €20, targeting the “University Low-Spenders” segment.

- `15% Discount on Combined Purchases of Energy Drinks and Protein Bars`: This promotion supports the “Tech-Gamer Customers” who tend to buy these products together.

- `10% Discount on Ratchet & Clank Bundle`: Customers buying Ratchet & Clank with Ratchet & Clank 2 get a 10% discount, targeting the “Big Family” cluster.

- `Energy Drink and Pancake Packs`: Bundled promotions for energy drinks and pancakes, reflecting popular joint purchases.

- `Loyalty Card Birthday Gift`: Customers holding a loyalty card receive a birthday cake as a special gift.
------

## **Achievement-Based Incentives**

To further motivate and reward customers, the following achievement-based incentives are suggested:

- `5% Discount on Electronics for Every 25 Energy Drinks Purchased`: Customers in the “Tech-Gamer” cluster receive a 5% discount on electronics each time their total energy drink purchases reach 25, accumulated over time.

- `Champagne Gift for Top Monthly Spenders`: Customers spending the most each month receive two bottles of champagne, rewarding the “High Spenders” cluster.

- `10% Discount on Games for Purchasing 20 Baby Foods`: Supports the “Big Family” cluster by incentivizing baby food purchases.

- `15% Discount on All Purchases After Buying 30 Bottles of Oil`: Encourages overall increased shopping frequency across all customer segments.

These tailored promotions leverage the insights from clustering and association rules, targeting specific customer behaviours to maximize engagement and sales growth.
